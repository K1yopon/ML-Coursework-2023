```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import (BatchNormalization, Conv2D,
                                     Conv2DTranspose, Dense, Dropout,
                                     Flatten, Input, LeakyReLU, Reshape)

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Create TensorFlow dataset from training data
x_dataset = tf.data.Dataset.from_tensor_slices(x_train)


def process_x(x):
    """Preprocesses the input image."""
    img = (tf.cast(x, tf.float32) - 127.5) / 127.5
    img = tf.expand_dims(img, axis=2)
    img = tf.image.resize(img, (32, 32))
    return img


# Apply preprocessing to the dataset
dataset = x_dataset.map(process_x)
dataset = dataset.shuffle(1000)

# Define batch size
batch_size = 256
dataset = dataset.batch(batch_size)

# Configure prefetching for performance
AUTOTUNE = tf.data.AUTOTUNE
dataset = dataset.cache().prefetch(buffer_size=AUTOTUNE)


def imshow():
    """Displays sample images from the dataset."""
    n = 10
    plt.figure(figsize=(10, 6))
    for images in dataset.take(1):
        for i in range(n):
            img = images[i]

            ax = plt.subplot(3, n, i + 1 + n)
            plt.imshow(img, cmap='gist_gray')
            plt.axis('off')
            ax.get_yaxis().set_visible(False)
        plt.show()


# Show sample images
imshow()

# Define Generator Model
generator_input = Input(shape=(128,))
x = Dense(8 * 8 * 256, use_bias=False)(generator_input)
x = BatchNormalization()(x)
x = LeakyReLU()(x)
x = Reshape((8, 8, 256))(x)
x = Conv2DTranspose(256, 3, strides=(1, 1), padding='same', use_bias=False)(
    x)
x = BatchNormalization()(x)
x = LeakyReLU()(x)
x = Conv2DTranspose(128, 3, strides=(2, 2), padding='same', use_bias=False)(
    x)
x = BatchNormalization()(x)
x = LeakyReLU()(x)
x = Conv2DTranspose(64, 3, strides=(2, 2), padding='same', use_bias=False)(x)
x = BatchNormalization()(x)
x = LeakyReLU()(x)
generator_output = Conv2DTranspose(
    1, 3, strides=(1, 1), padding='same', use_bias=False, activation='tanh')(
        x)
generator = keras.Model(generator_input, generator_output)

# Define Discriminator Model
discriminator_input = Input(shape=(32, 32, 1))
x = Conv2D(64, 3, strides=(2, 2), padding='same')(discriminator_input)
x = LeakyReLU()(x)
x = Dropout(0.3)(x)
x = Conv2D(128, 3, strides=(2, 2), padding='same')(x)
x = LeakyReLU()(x)
x = Dropout(0.3)(x)
x = Conv2D(128, 3, strides=(2, 2), padding='same')(x)
x = LeakyReLU()(x)
x = Dropout(0.3)(x)
x = Flatten()(x)
discriminator_output = Dense(1, activation='sigmoid')(x)
discriminator = keras.Model(discriminator_input, discriminator_output)


class GANModel(tf.keras.Model):
    """Defines the GAN model with generator and discriminator."""

    def __init__(self, generator_model, discriminator_model):
        super(GANModel, self).__init__()
        self.gen = generator_model
        self.disc = discriminator_model

        # Define optimizers for generator and discriminator
        self.gen_optimizer = tf.keras.optimizers.Adam(1e-4)
        self.disc_optimizer = tf.keras.optimizers.Adam(1e-4)

    def gan_loss(self, logits, is_real):
        """Calculates the GAN loss."""
        if is_real:
            labels = tf.ones_like(logits)
        else:
            labels = tf.zeros_like(logits)
        loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)(labels,
                                                                    logits)
        return loss

    @tf.function
    def training_step(self, x):
        """Performs a single training step."""
        noise = tf.random.normal([batch_size, 128], 0, 1, tf.float32)

        with tf.GradientTape() as disc_tape, tf.GradientTape() as gen_tape:
            fake_images = self.gen(noise, training=True)
            real_images = x

            fake_result = self.disc(fake_images, training=True)
            real_result = self.disc(real_images, training=True)

            disc_fake_loss = self.gan_loss(fake_result, False)
            disc_real_loss = self.gan_loss(real_result, True)

            disc_loss = disc_fake_loss + disc_real_loss

            gen_loss = self.gan_loss(fake_result, True)

        disc_gradients = disc_tape.gradient(disc_loss,
                                              self.disc.trainable_variables)
        self.disc_optimizer.apply_gradients(
            zip(disc_gradients, self.disc.trainable_variables))

        gen_gradients = gen_tape.gradient(gen_loss,
                                            self.gen.trainable_variables)
        self.gen_optimizer.apply_gradients(
            zip(gen_gradients, self.gen.trainable_variables))

        return tf.reduce_mean(disc_loss)


# Instantiate the GAN model
gan_model = GANModel(generator, discriminator)


def generate():
    """Generates and displays images from the generator."""
    n = 3
    plt.figure(figsize=(10, 6))

    for images in dataset.take(1):
        ims = images

        for i in range(n):
            img = gan_model.gen(tf.random.normal([1, 128], 0, 1,
                                                tf.float32))
            img = tf.squeeze(img, axis=0)
            ax = plt.subplot(3, n, i + 1)
            plt.imshow(img * 127.5 + 127.5, cmap='gist_gray')
            plt.title(
                gan_model.disc(tf.expand_dims(img, axis=0)).numpy()[0][0])
            plt.axis('off')

            ax = plt.subplot(3, n, i + 1 + n)
            plt.imshow(ims[i] * 127.5 + 127.5, cmap='gist_gray')
            plt.title(
                gan_model.disc(tf.expand_dims(ims[i], axis=0)).numpy()[0][0])
            ax.get_yaxis().set_visible(False)
            plt.axis('off')

        plt.show()


# Generate initial images
generate()

# Training loop
epochs = 50
from IPython.display import clear_output

loss_history = np.array([])  # Changed hist to loss_history
for epoch in range(1, epochs + 1):
    for step, x in enumerate(dataset):
        loss = gan_model.training_step(x)
        if step % 50 == 0:
            clear_output(wait=True)
            print(f"Epoch: {epoch}, Loss: {loss.numpy():.4f}")
            loss_history = np.append(loss_history, loss.numpy())  # Append loss
            plt.plot(np.arange(0, len(loss_history)), loss_history)  # Plot
            plt.xlabel("Iteration")  # Add labels
            plt.ylabel("Loss")
            plt.title("Discriminator Loss During Training")
            plt.show()
            generate()
```
